```python
import pandas as pd
url = "https://raw.githubusercontent.com/ageron/handson-ml/master/datasets/housing/housing.csv"
df = pd.read_csv(url)
#1.Count missing values in each column and drop rows with missing values.
missing_value = df.isnull().sum()
print(missing_value)
df_dropped_rows = df.dropna()
print(df_dropped_rows)
#2.Replace the missing values.
df_copy = df.copy()
df_copy['total_bedrooms'] = df_copy['total_bedrooms'].fillna(df_copy['total_bedrooms'].mean())
print(df_copy.isnull().sum())
#3.Filter rows where median income >5 and housing median age >30
income_df = df_copy[df_copy['median_income']>5]
print(income_df)
age_df = df_copy[df_copy['housing_median_age']>30]
print(age_df)
#4.Compute the average house value by ocean proximity
result = df.groupby('ocean_proximity')['median_house_value'].mean()
print(result)
#5.Sort dataset by median income(top 10 rows)
sort_df = df_copy.sort_values(by='median_income',ascending=False).head(10)
print(sort_df)
#6.Find correlation between median income and house value
correlation = df_copy['median_income'].corr(df_copy['median_house_value'])
print(correlation)
```
